```python
from crewai import Agent,Task,Crew,LLM
```

```python
from crewai.utilities.llm_utils import create_llm
```

```python
analyzer=Agent(
    role="AI Researcher",
    goal="Find useful and accurate information",
    backstory="Expert in AI and technology research",
    verbose=True,
    llm=create_llm(llm)
)
```

```python
# Research Agent
research_agent = Agent(
    role="AI Researcher",
    goal="Find useful and accurate information",
    backstory="Expert in AI and technology research",
    verbose=True,
    llm=create_llm(llm)
)
```

```python
# Writer Agent
writer_agent = Agent(
    role="Content Writer",
    goal="Write engaging and clear articles",
    backstory="Professional technical writer",
    verbose=True,
    llm=create_llm(llm)
)
```

```python
# Research Task
research_task = Task(
    description="""
    Research the latest AI trends in 2026.
    Focus on:
    - AI Agents
    - Robotics
    - Generative AI
    - Business Automation
    """,
    expected_output="Detailed research points about AI trends",
    agent=research_agent
)
```

```python
# Writing Task
writing_task = Task(
    description="""
    Using the research provided,
    write a professional article on AI trends in 2026.
    """,
    expected_output="Well-structured article",
    agent=writer_agent
)
```

```python
# Create Crew
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[research_task, writing_task],
    verbose=True
)
```

```python
# Run
result = crew.kickoff()

print("\nFINAL RESULT:\n")
print(result)
```
